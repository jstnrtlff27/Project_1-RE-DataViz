```python
import pandas as pd
import matplotlib.pyplot as plt

output_path="median price by state.csv"

state_data=pd.read_csv(output_path)
state_data

print("data supplied by 'https://www.realtor.com/research/data/'")
```

```python
state_data
```

```python
states_df=pd.DataFrame({
    "State": state_data["state"],
    "Average Listing Price": state_data["average_listing_price"],
    "Median Listing Price": state_data["median_listing_price"],
    "Median Listing Price per Square foot": state_data["median_listing_price_per_square_foot"]
})
states_df=states_df.sort_values(by="Average Listing Price", ascending=False)
states_df.plot(kind="bar", rot=45, figsize=(15, 10))
```

```python
import numpy as np
import matplotlib.pyplot as plt
states_df= states_df.sort_values(by="Median Listing Price", ascending=False)

x_axis=np.arange(len(states_df))
tick_locations=[value for value in x_axis]
plt.figure(figsize=(30, 15))
plt.bar(x_axis, states_df["Median Listing Price"])
plt.ylabel("US Dollars", fontsize=20)
plt.xlabel ("US States", fontsize=20)
plt.xticks(tick_locations, states_df.State.values, rotation=90, fontsize=15)
plt.yticks(fontsize=15)
plt.title("Median Listing Price by State", fontsize = 25)
plt.style.use('seaborn-poster')
```

```python
states_df=states_df.sort_values(by="Median Listing Price per Square foot", ascending=False)

x_axis=np.arange(len(states_df))
tick_locations=[value for value in x_axis]
plt.figure(figsize=(30, 15))
plt.bar(x_axis, states_df["Median Listing Price per Square foot"])
plt.xticks(tick_locations, states_df.State.values, rotation=90)
plt.xlabel("US States", fontsize=20)
plt.ylabel("US Dollars", fontsize=20)
plt.title("Median Listing Price per Square Foot by State", fontsize=25)
plt.show()
```

```python
#average rental price

rent_path="average rent price by state.csv"

rent_data=pd.read_csv(rent_path)
rent_data
```

```python
# merge states_df with rent_data
compare= pd.merge(states_df, rent_data, how="inner", on=["State", "State"])
compare
```

```python
compare=compare.sort_values(by="MedianRent", ascending=False)

x_axis=np.arange(len(compare))
tick_locations=[value for value in x_axis]
plt.figure(figsize=(30, 15))
plt.bar(x_axis, compare["MedianRent"])
plt.xticks(tick_locations, compare.State.values, rotation=90)
plt.xlabel("US States", fontsize=20)
plt.ylabel("US Dollars", fontsize=20)
plt.title("Median Rent Price per State", fontsize=25)
plt.show()
```
