```python
import pandas as pd
import matplotlib.pyplot as plt

path="Copy of RDC_Inventory_Core_Metrics_Zip.csv"

zip_data=pd.read_csv(path)

print("data supplied by 'https://www.realtor.com/research/data/'")
zip_data= pd.DataFrame(zip_data)

charleston=zip_data.loc[(zip_data["zip_name"]=="charleston, wv")]
# | (zip_data["postal_code"]==25302) | (zip_data["postal_code"]==25311)
#                        | (zip_data["postal_code"]==25306) | (zip_data["postal_code"]==25304) | (zip_data["postal_code"]==25314)
#                        | (zip_data["postal_code"]==25303) | (zip_data["postal_code"]==25309) | (zip_data["postal_code"]==25064)
#                        |(zip_data["postal_code"]==25313) | (zip_data["postal_code"]==25315) | (zip_data["postal_code"]==25177)]
charleston_price=pd.DataFrame({
    "Zip Code": charleston["postal_code"],
    "Average List price": charleston["average_listing_price"]
})
charleston_price
```

```python
charleston_merged=charleston_price.merge(charleston_df, how="left", on=["Zip Code", "Zip Code"])

charleston_merged= charleston_merged.loc[charleston_merged["Average Rent"]>0]
charleston_merged
```

```python
charleston_profit=pd.DataFrame({
    "Zip Code": charleston_merged["Zip Code"],
    "% Yearly Profit": (charleston_merged["Average Rent"]*12/charleston_merged["Average List price"])*100,
    "Average List Price": charleston_merged["Average List price"]
})
charleston_profit
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis=np.arange(len(charleston_profit))
tick_locations=[value for value in x_axis]
plt.figure(figsize=(7, 5))
plt.bar(x_axis, charleston_profit["% Yearly Profit"])
plt.xticks(tick_locations, charleston_profit["Zip Code"].values, rotation=45)
plt.title("Average Yearly profit by Zip Code")
plt.xlabel("Zip Code")
plt.ylabel("Average Yearly Profit (percentage)")
plt.show()

x_axis=np.arange(len(charleston_profit))
tick_locations=[value for value in x_axis]
plt.figure(figsize=(7, 5))
plt.bar(x_axis, charleston_profit["Average List Price"])
plt.xticks(tick_locations, charleston_profit["Zip Code"].values, rotation=45)
plt.title("Average List Price by Zip Code")
plt.xlabel("Zip Code")
plt.ylabel("Average List Price")
plt.show()
```
